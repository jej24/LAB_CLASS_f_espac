```
def display22(a,b,title1="Orginial", title2="Edited"):
    plt.figure(figsize=(10,10))
    plt.subplot(121), plt.imshow(a, cmap='gray'),plt.title(title1)
    plt.xticks([]),plt.yticks([])
    plt.subplot(122), plt.imshow(b, cmap='gray'), plt.title(title2)
    plt.xticks([]),plt.yticks([])
    plt.show()
    
```

```
def display4(a, b, c, d, title1= "Original", title2 = "Editado", title3 = "Editado1", title4 = "Editado2"):
   plt.figure(figsize=(20,20))
   plt.subplot(141), plt.imshow(a, cmap='gray'), plt.title(title1)
   plt.xticks([]), plt.yticks([])
   plt.subplot(142), plt.imshow(b, cmap='gray'), plt.title(title2)
   plt.xticks([]), plt.yticks([])
   plt.subplot(143), plt.imshow(c, cmap='gray'), plt.title(title3)
   plt.xticks([]), plt.yticks([])
   plt.subplot(144), plt.imshow(d, cmap='gray'), plt.title(title4)
   plt.xticks([]), plt.yticks([])
   plt.show()
```

```
def readRGB(imgPath):
    img = cv2.imread(imgPath)
    img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```
img = cv2.imread('INTER.jpg',0)
f = np.fft.fft2(img)
fshift = np.fft.fftshift(f)
magnitude_spectrum = 20*np.log(np.abs(fshift))


display22(img, magnitude_spectrum, "Input Image", "Magnitude Spectrum")
```

***FILTRO DE SHARPENING***

```
# Reading in and displaying our image
img = cv2.imread("INTER.jpg",0)

#create our sharpening kernel, it must equal to one eventually

kernel_sharpening = np.array([[-1,-1,-1],
                              [-1, 9,-1],
                              [-1,-1,-1]])

#applying the sharpening kernel to the input image & displaying it.

sharpened = cv2.filter2D(img, -1, kernel_sharpening)
display22(img, sharpened)
```

```
img = cv2.imread("saltpepper.png",0)

#apply the 3x3 median filter on the image

median = cv2.medianBlur(img, 3)
median = cv2.medianBlur(median, 3)
median = cv2.medianBlur(median, 3)
median = cv2.medianBlur(median, 5)
median = cv2.medianBlur(median, 5)
median = cv2.medianBlur(median, 5)
median = cv2.medianBlur(median, 7)

kernel_sharpening = np.array([[-1,-1,-1],
                              [-1, 9,-1],
                              [-1,-1,-1]])


#display image

sharpened = cv2.filter2D(median, -1, kernel_sharpening)
display22(img, sharpened)
```
